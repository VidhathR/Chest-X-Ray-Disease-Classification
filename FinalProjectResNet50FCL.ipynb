```python
from tensorflow import keras
datagen = keras.preprocessing.image.ImageDataGenerator(rescale=1 / 255.0,
                                                       vertical_flip = True,
                                                       horizontal_flip = True)
```

```python
from keras.applications import ResNet50

conv_base = ResNet50(weights = "imagenet",
include_top = True,
input_shape = (224, 224, 3))
conv_base.summary()
```

```python
trainGen = datagen.flow_from_directory(
    r'C:\Users\vidha\Health Analytics\archive\Covid19-dataset\train',
    target_size = (224,224),
    batch_size = 32,
    shuffle = True,
    class_mode = 'categorical'
)

datatestgen = keras.preprocessing.image.ImageDataGenerator(rescale=1 / 255.0)
                                                       
testGen = datatestgen.flow_from_directory(
    r'C:\Users\vidha\Health Analytics\archive\Covid19-dataset\test',
    target_size = (224,224),
    batch_size = 32,
    class_mode = 'categorical'
)
```

```python
model = keras.models.Model(inputs=conv_base.input, outputs=conv_base.layers[-2].output)
```

```python
x = model.output
x = keras.layers.Dense(3, activation = 'softmax')(x)
```

```python
new_model = keras.models.Model(inputs = model.input, outputs = x)
```

```python
for layer in new_model.layers[:-1]:
    layer.trainable = False
```

```python
new_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
new_model.fit(trainGen, epochs=10, batch_size=32, validation_data = testGen)
```

```python
new_model.save('ResNet50FinalProjectFCL.model')
```

```python
new_model.evaluate(testGen, batch_size = 32)
```
